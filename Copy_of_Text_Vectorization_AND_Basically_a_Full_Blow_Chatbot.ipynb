```python
import pandas as pd
```

```python
x1 = pd.DataFrame(["how are you?", "how are you? I", "how are you? I am", "how are you? I am doing", "how are you? I am doing well"])
```

```python
y = pd.DataFrame(["I", "am", "doing", "well", "<stop>"])
```

```python
my_le.classes_
```

```python
from sklearn.preprocessing import LabelEncoder


my_le = LabelEncoder()


encoded_y = my_le.fit_transform(y)
```

```python
from tensorflow.keras.layers import Dense, TextVectorization, Embedding, Flatten
```

```python
from tensorflow.keras.models import Sequential
```

```python
my_tv = TextVectorization(ragged = True)
```

```python
my_tv.adapt(x1)
```

```python
my_tv.get_vocabulary()
```

```python
vocab_size = len(my_tv.get_vocabulary())
```

```python
model = Sequential()
```

```python
model.add(Embedding(vocab_size, 3))

model.add(Flatten())

model.add(Dense(5, activation = 'softmax'))

model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam')
```

```python
input_sentence = "How are you?"
```

```python
my_tv(input_sentence)
```

```python
padded_sentence = pad_sequences([my_tv(input_sentence)], maxlen = 10)
```

```python
padded_sentence
```

```python
stop_token = "<stop>"

while True:

  output_probs = model.predict(padded_sentence, verbose = 0)

  # print(output_probs)

  output_index = output_probs.argmax()

  # print(output_index)

  output_text_set = my_le.inverse_transform([output_index])

  # print(output_text_set)

  true_output_text_token = output_text_set[0]

  # print(true_output_text_token)

  if true_output_text_token == stop_token:

    # print("stop token detected!")
    break #stop generating

  else:
    print(true_output_text_token, end = " ")

  internal_sentence_array = padded_sentence[0] #array of arrays -> 2D

  internal_sentence_list = internal_sentence_array.tolist() #convert to a list so that I can append it

  my_encoded_token = my_tv(true_output_text_token) #can't append the text, I need a number, so I need to convert the text to the proper number

  my_integer = my_encoded_token.numpy().tolist()[0]

  internal_sentence_list.append(my_integer)

  padded_sentence = pad_sequences([internal_sentence_list], maxlen = 10)
```
